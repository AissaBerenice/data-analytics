```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

housingdata = pd.read_csv("../data/raw/BostonHousing.csv")
```

```python
X= np.array(housingdata['lstat']).reshape(-1,1)
y= np.array(housingdata['medv']).reshape(-1,1)

modelo=LinearRegression()
modelo.fit(X,y)
y_pred2=modelo.predict(X)
```

```python
plt.figure(figsize=(10,4))


plt.scatter(X,y, label='Data')


plt.plot(X, y_pred2, label='Linear Regression',c='green')

plt.xlabel('lstat')
plt.ylabel('medv')
plt.title('Aissa Berenice Gonzalez Fosado')
plt.legend()
```
