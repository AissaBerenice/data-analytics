```python
#Estadisticas Descriptivas de uno de los features

mean1 = df["Deaths - Meningitis - Sex: Both - Age: All Ages (Number)"].mean()
print(f'The Average for Meningitis Deaths is of  {mean1}')

median1 = df["Deaths - Meningitis - Sex: Both - Age: All Ages (Number)"].median()
print(f'The Median for Meningitis Deaths is of  {median1}')

mode1 = df["Deaths - Meningitis - Sex: Both - Age: All Ages (Number)"].mode()
print(f'The Median for Meningitis Deaths is of  {mode1}')
```
