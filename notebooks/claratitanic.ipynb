```python
# Importar librerias y documentos
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv("../data/raw/tested.csv")


# Asignar Passenger Id como indice
df = df.set_index("PassengerId")
df.head()
```

1. ¿Cuantos pasajeros murieron y sobrevivieron?


```python
sobrevivieron = df["Survived"].value_counts()[1]
murieron = df["Survived"].value_counts()[0]
print(f"sobrevivieron {sobrevivieron} pasajeros y murieron {murieron}, pasajeros")
```

2. ¿Hubo alguien que pago demasioado por el boleto?


```python
#Rango intercuartilico
IQR = df.Fare.quantile(0.75) - df.Fare.quantile(0.75)
print(IQR)

#Limites 
lower = df.Fare.quantile(0.25) - 1.5 * IQR
upper = df.Fare.quantile(0.75) - 1.5 * IQR
print(f"lower: {lower} upper: {upper}")

#Outliers, Datos fuera del lower and upper limits
outliers = df["Fare"]
for i in outliers:
    if i < upper:
        print(i,outliers)
outliers.__len__
```

3. ¿Cual es la media, mediana y moda del costo del ticket?


```python
media = df["Fare"].mean()
mediana = df["Fare"].median()
moda = df["Fare"].mode()
print(f"La media es {media}, la mediana {mediana} y la moda {moda}")
```

4. ¿Cual es la varianza de las edades de los pasajeros?

```python
#Varianza
var_edades = df["Fare"].var()
print(var_edades)
```

5. ¿Quien es mas probable que sobreviva: Una mujer de 40 años o un hombre de 70 años?


```python
#probabilidad de una mujer de 40 anos
df_pro=df[['Survived','Age','Sex']]
df_pro

df_pro['Age'].fillna(df_pro['Age'].median(), inplace=True)
num_personas=len(df)
personas_sobrevivientes=df_pro['Survived'].sum()

df_pro_mujeres=df_pro[df_pro['Sex']=='female']
print(df_pro_mujeres)

num_mujeres=len(df_pro_mujeres)
print(f'El número de mujeres en el barco era de: ',num_mujeres)
edades_female=df_pro_mujeres['Age'].min(),df_pro_mujeres['Age'].max()
print(f'Las edades máximas y mínimas de la mujer eran:', edades_female)

edad=0
for i in range(10,80,10):
    #print(edad,i)
    personas_mujeres = df_pro_mujeres[(df_pro_mujeres['Age']>=edad)&(df_pro_mujeres['Age']<i)]
    plt.figure()
    n= plt.hist(personas_mujeres['Survived'], density=True)
    plt.title('mujeres survived en el rango de edad:'+str(edad)+','+str(i))
    edad=i
    print(n)

#hombres 
df_pro_hombres=df_pro[df_pro['Sex']=='male']
df_pro_hombres

num_hombres=len(df_pro_hombres)
print(f'El número de hombres en el barco era de: ',num_hombres)
edades= df_pro_hombres['Age'].min(),df_pro_hombres['Age'].max() 
print(f'La edad máxima y mínima del hombre era: ',edades)
edad=0
for i in range(10,90,10):
    #print(edad,i)
    personas_hombres = df_pro_hombres[(df_pro_hombres['Age']>=edad)&(df_pro_hombres['Age']<i)]
    plt.figure()
    n=plt.hist(personas_hombres['Survived'], density=True)
    plt.title('hombres survived en el rango de edad:'+str(edad)+','+str(i))
    edad=i
    print(n)
```
