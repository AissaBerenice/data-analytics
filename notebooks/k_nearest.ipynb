```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
X,y = make_blobs(n_samples = 100, n_features= 2, centers = 2, random_state= 3, cluster_std= 5)
X.shape, y.shape
```

```python
#Primera Dimension
X[:,0]
```

```python
plt.scatter(X[:,0], X[:,1], c= y == 0, label = 'Class 0')
plt.scatter(X[:,0], X[:,1], c= y == 1, label = 'Class 0')


plt.plot()
```

## Model

```python
#Importar Clase
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Crear un objeto de una clase El objeto es la instancia de una clase
clf = KNeighborsClassifier()
```

```python
clf.fit(X,y)
```

```python
clf.score(X,y)
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test =train_test_split(X,y, test_size= 0.2, random_state= 3)
X_train.shape, X_test.shape
```

```python
clf.fit(X_train,y_train)
clf.fit(X_test,y_test)
```
